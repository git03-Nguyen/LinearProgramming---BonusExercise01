>Một nghiên cứu cho thấy tác động của vitamin A, B đối với cơ thể người trong một ngày, cụ thể là:
>- Cơ thể tiếp nhận không quá 600 đơn vị vitamin A và không quá 500 đơn vị vitamin B.
>- Cơ thể tiếp nhận tốt nhất từ 400 đến 1000 đơn vị vitamin cả A lẫn B. 
>- Do tác động phối hợp, số đơn vị vitamin B không ít hơn nửa số đơn vị vitamin A nhưng cũng không nhiều hơn 3 lần đơn vị vitamin A.
>
>Biết rằng giá của một đơn vị vitamin A là 9 đồng, còn của 1 đơn vị vitamin B là 7,5 đồng. Tính chi phí ít nhất và nhiều nhất mỗi ngày cần để có thể dùng đủ cả hai loại vitamin.\

## <span style="color:green">Bài làm:</span>
Gọi $x$, $y$ lần lượt là số đơn vị vitamin A và vitamin B mà người sử dụng trong một ngày.\
Theo đề bài, ta có các ràng buộc sau:
$$
\begin{cases}
    x, y \in N^* \\
    x \le 600 \\
    y \le 500 \\
    400 \le x + y \le 1000 \\
    \frac{1}{2}x \le y \le 3x
\end{cases}
\quad
\Leftrightarrow 
\quad
\begin{cases}
    x, y \in N \\
    1 \le x \le 600 \\
    1 \le y \le 500 \\
    400 \le x + y \le 1000 \\
    \frac{1}{2}x \le y \le 3x
\end{cases}
$$

Hàm mục tiêu $F(x,y)=9x+7.5y$ biểu diễn chi phí để sử dụng vitamin. \
Từ các ràng buộc, ta tiến hành vẽ các đường thẳng lên hệ trục toạ độ $Oxy$ như bên dưới:
$
\begin{cases}
    x = 1 \\
    y = 1 \\
    x = 600 \\
    y = 500 \\
    x + y = 400 \\
    x + y = 1000 \\
    \frac{1}{2}x = y \\
    y = 3x
\end{cases}
$

Dựa vào đồ thị ta thấy miền ràng buộc bị chặn, và các điểm cực biên là: 
$$
\begin{equation}
\begin{split}
    A(\frac{500}{3}, 500) \rightarrow \text{Loại} \quad &; \quad B(500,500) \rightarrow F(500,500) = 8250 \quad &; \quad \boldsymbol{C(600, 400) \rightarrow F(600,400) = 8400} \\
    D(600,300) \rightarrow F(600,300) = 7650 \quad &; \quad E(\frac{800}{3}, \frac{400}{3}) \rightarrow \text{Loại} \quad &; \quad \boldsymbol{F(100,300) \rightarrow F(100,300)=3150 }
\end{split}
\end{equation}
$$

Vì miền ràng buộc bị chặn, suy ra GTNN là $F(100,300)=3150$ và GTLN là $F(600,400)=8400$.

Vậy **chi phí ít nhất là $3150$ đồng** khi sử dụng 100 đơn vị vitamin A và 300 đơn vị vitamin B. \

Và **chi phí nhiều nhất là $8400$ đồng** khi sử dụng 600 đơn vị vitamin A và 400 đơn vị vitamin B.

## <span style="color:green">2.1. Ý tưởng:</span>
- Sử dụng thư viện *numpy* cho các công việc tích hai ma trận, giải hệ phương trình bậc nhất hai ẩn.
- Nhập và lưu trữ các ràng buộc dưới dạng ma trận các hệ số.
- Nhập và lưu trữ hàm mục tiêu bằng vector pháp tuyến của nó.
- **Lưu ý**: 
    - Có kiểm tra điều kiện đầu vào như đề bài đã cho.
    - Có lưu ý đến sai số floating-point khi so sánh "bằng" (sử dụng $\varepsilon = 10^{-9}$).
    - Khi xảy ra ngoại lệ, lập tức dừng chương trình và báo lỗi sử dụng *try..except*.
- **Ý tưởng ý 1**:
    - Coi danh sách các ràng buộc là những đường thẳng.
    - Tìm tất cả giao điểm giữa tất cả các cặp trong danh sách đường thẳng.
    - Từ danh sách giao điểm, lọc ra những điểm thoả mãn tất cả ràng buộc $\Rightarrow$ **Danh sách điểm cực biên**.
- **Ý tưởng ý 2**:
    - Từ danh sách các điểm cực biên, tìm một bao lồi của chúng sử dụng thuật toán "Chuỗi đơn điệu" (Monotone chain algorithm).
    - Kiểm tra có cạnh bao lồi nào KHÔNG nằm trên ít nhất một đường ràng buộc hay không, bằng cách: \
    Xét tất cả trung điểm từng cạnh bao lồi:
        - Nếu có bất cứ trung điểm nào mà: KHÔNG nằm trên bất cứ đường ràng buộc nào $\Rightarrow$ **Miền ràng buộc không bị chặn**.
        - Nếu tất cả trung điểm đều: nằm trên ít nhất 1 đường ràng buộc $\Rightarrow$ **Miền ràng buộc bị chặn**.
- **Ý tưởng ý 3**:
    - Nếu miền ràng buộc bị chặn: **GTLN và GTNN là tồn tại**, đạt tại hai trong những điểm cực biên.
    - Nếu miền ràng buộc không bị chặn:
        - Tìm vector pháp tuyến $\vec{n}$ của đường thẳng biểu diễn hàm mục tiêu $F: ax + by$.
        - Từ điểm cực biên tạo ra GTLN, ta di chuyển theo vector $\vec{n}$, nếu tồn tại bất kỳ điểm $M(x_M,y_M)$ nào mà $F(M) > \text{GTLN}$ và thoả mãn tất cả ràng buộc (tức $M$ nằm trong miền ràng buộc) thì khi đó **GTLN là không tồn tại**. Nếu không có điểm $M$ nào như vậy thì **GTLN đạt tại một trong những cực biên**.
        - Tương tự với GTNN, nhưng lần này ta bắt đầu từ điểm cực biên tạo ra GTNN, di chuyển theo vector $-\vec{n}$, nếu tồn tại bất kỳ điểm $N(x_N,y_N)$ nào mà $F(N) < \text{GTNN}$ và thoả mãn tất cả ràng buộc (tức $N$ nằm trong miền ràng buộc) thì khi đó **GTNN là không tồn tại**. Nếu không có điểm $N$ nào như vậy thì **GTNN đạt tại một trong những cực biên**.


## <span style="color:green">2.2. Khai báo hàm hỗ trợ:</span>
>Dựa trên ý tưởng, tiến hành cài đặt hàm hỗ trợ.\
>Giải thích chi tiết về cài đặt đã được comment cẩn thận vào trong mã nguồn.

```python
import numpy as np
# ----------------------------------------------------
# Hàm Nhập điều kiện bài toán: các ràng buộc và hàm mục tiêu
def input_problem():
    _m = 0
    _constraints = []
    _obj_func = []
    
    # Nhập số ràng buộc
    _m = int(input())
    if _m <= 0 or _m > 10: raise Exception("Số ràng buộc không hợp lệ!")
        
    # Nhập các ràng buộc dạng ax + by <= c
    for i in range(_m):
        s = input().split()
        s[:3] = [float(x) for x in s[:3]]            
        _constraints.append(s)
    _constraints.append([-1., 0., 0.])
    _constraints.append([0., -1., 0.])

    # Nhập hàm mục tiêu dạng F(x, y) = ax + by
    _obj_func = input().split()
    _obj_func[:2] = [float(x) for x in _obj_func[:2]] 
    
    return _constraints, _obj_func

# ----------------------------------------------------
# Hàm tìm danh sách các điểm giao nhau
def find_intersect_pts(constraints):
    constraints = np.array(constraints)
    _intersect_pts = []
    n = len(constraints)
    
    for i in range(n - 1):
        for j in range(i + 1, n):
            
            eqn_1 = constraints[i]
            eqn_2 = constraints[j]
            
            coefficients = np.vstack((eqn_1[:2], eqn_2[:2]))
            constants = np.array([eqn_1[2], eqn_2[2]])
            
            try:
                # Giải hệ pt
                solution = np.linalg.solve(coefficients, constants)
                # Thêm vào danh sách điểm giao nhau
                _intersect_pts.append(solution)
                
            except np.linalg.LinAlgError:
                # Nếu hai đường thằng song song
                continue
                
    return _intersect_pts

# ----------------------------------------------------
# Hàm tìm các điểm cực biên từ các điểm giao nhau mà thoả mãn tất cả ràng buộc
def find_satisfied_pts(constraints, intersect_pts):
    constraints = np.array(constraints)
    extreme_pts = np.array(intersect_pts)

    dot_product = np.dot(constraints[:, :2], extreme_pts.T)

    # Epsilon cho sai số floating-point number
    epsilon = 1e-9
    
    # Lọc ra các điểm thoả mãn tất cả các ràng buộc
    mask = np.all(dot_product <= constraints[:, 2][:, np.newaxis] + epsilon, axis=0)
    extreme_pts_filtered = extreme_pts[mask]
    
    return extreme_pts_filtered

# ----------------------------------------------------
# Hàm in danh sách các điểm
def print_points(points_2D):
    str = ''
    length = len(points_2D)
    for i in range(length):
        [x, y] = points_2D[i]
        str += f'({x}, {y})'
        if i < length - 1: str += ', '
    return str
    
# ----------------------------------------------------
# ----------------------------------------------------
# Hàm tính toán tập bao lồi sử dụng thuật toán "Chuỗi đơn điệu" (Monotone chain algorithm)
def convex_hull(points):
    n = len(points)
    if (n < 3): 
        raise Exception('Cần ít nhất 3 điểm để tìm được bao lồi!')
        
    hull = []
    
    # Tìm điểm trái nhất
    leftmost = 0
    for i in range(1, n):
        if (points[i][0] < points[leftmost][0]):
            leftmost = i

    # Bắt đầu từ điểm trái nhất, di chuyển ngược chiều kim đồng hồ
    # đến khi gặp điểm bắt đầu
    p = leftmost
    q = 0
    while (True):
        # Thêm điểm hiện tại vào bao
        hull.append(points[p])
        
        # Tìm kiếm điểm q sao cho orientation(p, x, q) 
        # là ngược chiều kim đồng hồ với mọi điểm x. 
        q = (p + 1) % n
        for i in range(0, n):
            # Nếu có bất cứ điểm i nào ngược chiều kim đồng hồ
            # hơn là q, thì cập nhật q
            if (orientation(points[p], points[i], points[q]) == 2):
                q = i
                
        # Bây giờ q là điểm ngược chiều kim đồng hồ nhất so với p.
        # Thêm q vào bao và chọn p = q cho vòng lặp tiếp theo
        p = q
 
        # Tiếp tục cho đến khi đến điểm bắt đầu thì dừng
        if (p == leftmost):
            break
 
    # Kết quả bao lồi
    return hull
    
# ----------------------------------------------------
# Hàm tính toán hướng sắp xếp của bộ 3 điểm (p, q, r) là:
# 0 -> Thẳng hàng
# 1 -> Chiều kim đồng hồ
# 2 -> Ngược chiều kim đồng hồ
def orientation(p, q, r):
    # Tính định thức của các vector pq và qr
    val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])

    if (val == 0):
        return 0  
    elif (val > 0):
        return 1  
    else:
        return 2 
# ----------------------------------------------------
# Hàm tìm trung điểm
def midpoint(p1, p2):
    x = (p1[0] + p2[0]) / 2
    y = (p1[1] + p2[1]) / 2
    return [x, y]
# ----------------------------------------------------
# Hàm hỗ trợ so sánh tránh sai số floating-point number
def is_equal(num1, num2, epsilon = 1e-9):
    return (num1 >= (num2 - epsilon)) and (num1 <= (num2 + epsilon))
# ----------------------------------------------------
# Hàm kiểm tra miền ràng buộc có bị chặn hay không
# Ý tưởng là các cạnh của bao lồi có trùng với ít nhất một đường ranh giới ràng buộc hay không
# TƯơng đương với việc là sẽ thử lần lượt trung điểm của cạnh bao lồi vào lần lượt các đường ràng buộc
def check_bounded_feasible(constraints, hull):
    
    # Tính toán danh sách các trung điểm cạnh bao lồi
    midpoints = []
    for i in range(len(hull)):
        next_index = (i + 1) % len(hull)
        mid = midpoint(hull[i], hull[next_index])
        midpoints.append(mid)
        
    # Với mỗi điểm
    for point in midpoints:
        bounded = True
        # Thế vào lần lượt từng đường ràng buộc ax + by = c
        for constraint in constraints:
            left = constraint[0] * point[0] + constraint[1] * point[1]
            right = constraint[2]
            # Nếu nằm trên ít nhất một đường ràng buộc thì xét điểm (cạnh bao lồi) tiếp theo.
            if is_equal(left, right):
                bounded = True
                break
            else: 
                bounded = False
        # Nếu tồn tại một điểm mà không nằm trên bất cứ đường ràng buộc nào thì kết luận miền rằng buộc không bị chặn
        if not bounded:
            return False
    
    return True

# ----------------------------------------------------
# Hàm tìm điểm (x,y) tạo nên GTLN cho func: ax + by
def find_max(func, points):
    max_val = float('-inf')
    max_point = None
    for point in points:
        value = func[0] * point[0] + func[1] * point[1]  # ax + by
        if value > max_val:
            max_val = value
            max_point = point
    
    return max_val, max_point[0], max_point[1]
# ----------------------------------------------------
# Hàm tìm điểm (x,y) tạo nên GTNN cho func: ax + by
def find_min(func, points):
    min_val = float('+inf')
    min_point = None
    for point in points:
        value = func[0] * point[0] + func[1] * point[1]  # ax + by
        if value < min_val:
            min_val = value
            min_point = point
    
    return min_val, min_point[0], min_point[1]
# ----------------------------------------------------
# ----------------------------------------------------
# ---------------- MAIN FUNCTION ---------------------
# ----------------------------------------------------

# Sau khi khởi tạo các hàm hỗ trợ, dưới đây là hàm chính để giải quyết bài toán
def solve_problem():
    try:
        # Nhập bài toán
        constraints, obj_func = input_problem()

        # Tìm danh sách các điểm giao nhau
        intersect_pts = find_intersect_pts(constraints)
        # Tìm các điểm cực biên từ các điểm giao nhau mà thoả mãn tất cả ràng buộc
        extreme_pts = find_satisfied_pts(constraints, intersect_pts)

        # 1) In ra các điểm cực biên
        print(f'1) Danh sach cac diem cuc bien la: {print_points(extreme_pts)}.')

        # Tìm bao lồi của các điểm cực biên
        hull = convex_hull(extreme_pts)

        # 2) In ra kết luận miền ràng buộc bị chặn hay không
        is_bounded = check_bounded_feasible(constraints, hull)
        if is_bounded: 
            print('2) Mien rang buoc la bi chan.')
        else:
            print('2) Mien rang buoc khong bi chan.')

        # Tìm GTLN và GTNN trong ds cực biên
        max_val, max_x, max_y = find_max(obj_func, extreme_pts)
        min_val, min_x, min_y = find_min(obj_func, extreme_pts)

        # 3) TH1: Nếu miền rb bị chặn thì kết luận GTLN và GTNN
        if is_bounded:
            print(f'3) GTLN la F = {max_val} tai x1 = {max_x}, x2 = {max_y}.')
            print(f'GTNN la F = {min_val} tai x1 = {min_x}, x2 = {min_y}.')

        # 3) TH2: Nếu miền rb không bị chặn thì:
        else:
            # Với F(x,y) = ax + by => Tìm vector pháp tuyến dương của  F chính là (a, b) và pháp tuyến âm là (-a, -b)
            n = obj_func

            # Xem xét GTLN: đi từ GTLN của cực biên, nếu men theo pháp tuyến dương mà 
            # còn giá trị nằm trong miền rb và lớn hơn GTLN thì không tồn tại GTLN (chứng minh quy nạp)
            new_max = [max_x + n[0], max_y + n[1]]
            new_max_val = obj_func[0] * new_max[0] + obj_func[1] * new_max[1]  # ax + by
            res_max = find_satisfied_pts(constraints, [new_max])
            if len(res_max) > 0 and new_max_val > max_val:
                print(f'3) GTLN khong ton tai.')
            else:
                print(f'3) GTLN la F = {max_val} tai x1 = {max_x}, x2 = {max_y}.')

            # Tương tự, xem xét GTNN: đi từ GTNN của cực biên, nếu men theo pháp tuyến âm mà 
            # còn giá trị nằm trong miền rb và nhỏ hơn GTNN thì không tồn tại GTNN (chứng minh quy nạp)
            new_min = [min_x - n[0], min_y - n[1]]
            new_min_val = obj_func[0] * new_min[0] + obj_func[1] * new_min[1]  # ax + by
            res_min = find_satisfied_pts(constraints, [new_min])
            if len(res_min) > 0 and new_min_val < min_val:
                print(f'GTNN khong ton tai.')
            else:
                print(f'GTNN la F = {min_val} tai x1 = {min_x}, x2 = {min_y}.')

    except Exception as e:
        print("ERROR: ", e)
        
# ----------------------------------------------------
    
```
